```python
from sklearn.model_selection import train_test_split

X = np.array([df[column].values for column in df.columns if column != 'target']).transpose()
Y = data.target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
params = {'legend.fontsize': 'large',
          'figure.figsize': (12, 8),
          'figure.titlesize': 'x-large',
          'axes.labelsize': 'x-large',
          'axes.titlesize':'x-large',
          'xtick.labelsize':'x-large',
          'ytick.labelsize':'x-large',
          'lines.markersize': 10
          }
pylab.rcParams.update(params)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import BernoulliNB

X = np.array([df[column].values for column in df.columns if column != 'target']).transpose()
Y = data.target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
Y_pred = bnb.predict(X_test)
plt.scatter(0.2, (Y_test != Y_pred).sum(), 50, c = '#1f77b4', label = f'Size/Error observations')

change_bounds = np.arange(0.05, 0.51, 0.05, dtype = float)
size_errors = []
for i in change_bounds:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=i, random_state=50)
    bnb = BernoulliNB().fit(X_train, Y_train)
    Y_pred = bnb.predict(X_test)
    size_errors.append([i, (Y_test != Y_pred).sum()])
    plt.scatter(i, (Y_test != Y_pred).sum(), 50, c = '#1f77b4')

size_errors = np.array(size_errors)

x_Err = np.array([[size_errors[i,0]] for i in range(len(size_errors))])
reg = LinearRegression().fit(x_Err, size_errors[:,1])

plt.plot(size_errors[:,0], reg.predict(x_Err), label = f'Size/Error dependence '
                                                           f'(y={np.round(reg.coef_[0],2)}x + {np.round(reg.intercept_,2)})')

plt.xlabel('Test set size')
plt.ylabel('Number of misclassified examples')
plt.xlim(0,0.52)
plt.legend()
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(bnb.predict(X_test), Y_test, normalize='true'),
            annot = True, annot_kws={'fontsize': 16})
```
