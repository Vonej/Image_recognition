```python
import pandas as pd
import numpy as np
df=pd.read_csv('car_data.csv')
df=df.drop(columns=['User ID', 'Gender', 'Purchased'])
df
```

```python
import matplotlib.pylab as pylab
import matplotlib.pyplot as plt
params = {'legend.fontsize': 'large',
          'figure.figsize': (12, 10),
          'figure.titlesize': 'x-large',
          'axes.labelsize': 'x-large',
          'axes.titlesize':'x-large',
          'xtick.labelsize':'x-large',
          'ytick.labelsize':'x-large',
          'lines.markersize': 10
          }
pylab.rcParams.update(params)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.4, wspace=0.4)
for i, column in enumerate(df.columns, 1):
    plt.subplot(2,1,i)
    sns.histplot(df[column])
```

```python
df.describe()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import normalize
data = df[['Age','AnnualSalary']].to_numpy(dtype='float')
scaler = preprocessing.Normalizer().fit(data[:200,:])
data_scaled = scaler.transform(data)
normalized_df = df.copy()
normalized_df[['Age','AnnualSalary']] = data_scaled
normalized_df
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import normalize
normalized_df = df[:250].copy()

for column in df.columns:
    normalized_df[column] = normalize(df[column][:250,np.newaxis], axis=0).ravel()

```

```python
plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.4, wspace=0.4)
for i, column in enumerate(normalized_df.columns, 1):
    plt.subplot(2,1,i)
    sns.histplot(normalized_df[column])
```
